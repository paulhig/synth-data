```
import sklearn.datasets
import pylab
import numpy
#Specifiy the number of examples we need (500) and the noise level
train_X, train_y = sklearn.datasets.make_moons(500, noise=0.1)
#One hot encode the target values
train_y_onehot = numpy.eye(2)[train_y]
#Plot the data
pylab.scatter(train_X[:-100, 0], train_X[:-100, 1], c=train_y[:-100], cmap=pylab.cm.Spectral)
```
